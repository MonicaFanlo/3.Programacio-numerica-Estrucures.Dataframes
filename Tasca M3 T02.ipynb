```python
delay['Speed'] = delay['Distance']/delay['AirTime'] #CÃ lcul de la velocitat de vol
```

```python
# es consedera un gran retard a partir de 180 min i s'han de pagar compensacions als usuaris
def estat_function(x):
    if x <= 15:
        return 'On-Time'
    elif x > 15 and x <=180:
        return 'Minor Delay'
    elif x > 180:
        return 'Great Delay'
          

delay['Estat'] = delay['ArrDelay'].apply(estat_function) #CÃ lcul per determinar el tipus de retard
```

#### Per cÃ lculs endarreriments

```python
not_cancelled = delay['Cancelled']==0
not_diverted = delay['Diverted'] ==0
delayed = delay[not_cancelled & not_diverted] #NomÃ©s agafem els vols que no han estat cancelats ni derivats
```

```python
grouped=delayed.groupby('UniqueCarrier')
grouped['ArrDelay'].agg(np.sum).sort_values(ascending=False).head(10) #Les 10 companyies amb mÃ©s retards acumulats
```

```python
groupedvols=delayed.groupby(['Origin','Dest'])
groupedvols['AirTime'].sum().sort_values(ascending=False).head(10)  #10 rutes mÃ©s llargues
```

```python
groupedvols['ArrDelay'].sum().sort_values(ascending=False).head(10) # 10 rutes que acumulen mÃ©s retards
```

## Altres cÃ lculs

```python
#CÃ lcul per determinar l'importÃ ncia de vols cancelÂ·lats o desviats. Segur que hi ha un altra manera mÃ©s curta i eficient de fer-ho
#ho he provat de diferents maneres perÃ² no he estat capaÃ§ de solucionar-ho amb una funciÃ³ que tingues en compte dos columnes    
def status(x):
    if x == '0-0':
        return 'opertated'
    elif x =='0-1':
        return 'diverted'
    elif x =='1-0':
        return 'cancelled'

delay[['Cancelled','Diverted']]=delay[['Cancelled','Diverted']].astype(str)
delay['Status']= delay[['Cancelled','Diverted']].apply('-'.join,axis=1) 
delay['Status'] = delay['Status'].apply(status)
delay['Status'].value_counts()/len(delay['Status'])
```

```python
#Retard mitjÃ  dels vols per mes

delayed[['Month','ArrDelay']].groupby(['Month']).mean().plot()
plt.show()
```

```python
#per veure com afecta els deferents tipus d'incicÃ¨ncies als endarreriments per mesos
parameters = ['CarrierDelay','WeatherDelay','NASDelay','SecurityDelay','LateAircraftDelay']
delayed[['Month','CarrierDelay','WeatherDelay','NASDelay','SecurityDelay','LateAircraftDelay']].groupby('Month')[parameters].sum().plot()
```

```python
# retard mitjÃ  dels vols per dia de la setmana
delayed[['DayOfWeek','ArrDelay']].groupby(['DayOfWeek']).mean().plot()
plt.show()
```
