#### Variables del dataset
* Unnamed:0 : Es una variable d'index del dataset
* Year : any de les dades (2008)
* Month : mes
* DayofMonth : dia
* DayofWeek : dia de la setmana 1 (dilluns)-7(diumenge)
* DepTime : hora real que surt l'avió de la porta (hhmm)
* CRSDepTime : hora programada de sortida de l'avió (hhmm)
* ArrTime : hora real que arriba l'avió a la porta (hhmm)
* CRSArrTime : hora programada que arriba l'avió (hhmm)
* UniqueCarrier : companyia aaèria
* FlightNum :número de vol
* TailNum : número de l'avió
* ActualElapsedTime : duració del vol real (entre portes d'arribada i sortida) (minuts)
* CRSElapsedTime : duració teòrica del vol (entre portes d'arribada i sortida) (minuts)
* AirTime : temps de vol (minuts)
* ArrDelay : temps de retard en l'arribada (on-time quan és menys de 15 min de l'arribada programada) (minuts)
* DepDelay : temps de retard en la sortida (minuts)
* Origin : ciutat d'origin (minuts)
* Dest : ciutat de destinació
* Distance : distància (milles)
* TaxiIn : temps des de l'aterratge fins arribar a portes (minuts)
* TaxiOut : temps des de portes fins a l'enlairament (minuts)
* Cancelled : si el vol ha estat cancel·lat o no (0:NO cancel·lat / 1: Cancel·lat)
* CancellationCode : causa de la cancel·lació (N : companyia / B : temps / A : NAS / C : seguretat)
* Diverted : si el vol ha estat desviat (0: No desviat / 1:desviat)
* CarrierDelay : retard a causa de la companyia (minuts)
* WheatherDelay : retard a causa del temps (minuts)
* NASDelay : retard a causa del sistema nacional aeroespacial (minuts)
* SegurityDelay : retard a causa de la seguretat (minuts)
* LateAircraftDelay : retard a causa de l'arribada amb retard de l'anterior vol (minuts)

```python
delay['Speed'] = delay['Distance']/delay['AirTime'] #Càlcul de la velocitat de vol
```

```python
# es consedera un gran retard a partir de 180 min i s'han de pagar compensacions als usuaris
def estat_function(x):
    if x <= 15:
        return 'On-Time'
    elif x > 15 and x <=180:
        return 'Minor Delay'
    elif x > 180:
        return 'Great Delay'
          

delay['Estat'] = delay['ArrDelay'].apply(estat_function) #Càlcul per determinar el tipus de retard
```

#### Per càlculs endarreriments

```python
not_cancelled = delay['Cancelled']==0
not_diverted = delay['Diverted'] ==0
delayed = delay[not_cancelled & not_diverted] #Només agafem els vols que no han estat cancelats ni derivats
```

```python
grouped=delayed.groupby('UniqueCarrier')
grouped['ArrDelay'].agg(np.sum).sort_values(ascending=False).head(10) #Les 10 companyies amb més retards acumulats
```

```python
groupedvols=delayed.groupby(['Origin','Dest'])
groupedvols['AirTime'].sum().sort_values(ascending=False).head(10)  #10 rutes més llargues
```

```python
groupedvols['ArrDelay'].sum().sort_values(ascending=False).head(10) # 10 rutes que acumulen més retards
```

## Altres càlculs

```python
#Càlcul per determinar l'importància de vols cancel·lats o desviats. Segur que hi ha un altra manera més curta i eficient de fer-ho
#ho he provat de diferents maneres però no he estat capaç de solucionar-ho amb una funció que tingues en compte dos columnes    
def status(x):
    if x == '0-0':
        return 'opertated'
    elif x =='0-1':
        return 'diverted'
    elif x =='1-0':
        return 'cancelled'

delay[['Cancelled','Diverted']]=delay[['Cancelled','Diverted']].astype(str)
delay['Status']= delay[['Cancelled','Diverted']].apply('-'.join,axis=1) 
delay['Status'] = delay['Status'].apply(status)
delay['Status'].value_counts()/len(delay['Status'])
```

```python
#Retard mitjà dels vols per mes

delayed[['Month','ArrDelay']].groupby(['Month']).mean().plot()
plt.show()
```

```python
#per veure com afecta els deferents tipus d'incicències als endarreriments per mesos
parameters = ['CarrierDelay','WeatherDelay','NASDelay','SecurityDelay','LateAircraftDelay']
delayed[['Month','CarrierDelay','WeatherDelay','NASDelay','SecurityDelay','LateAircraftDelay']].groupby('Month')[parameters].sum().plot()
```

```python
# retard mitjà dels vols per dia de la setmana
delayed[['DayOfWeek','ArrDelay']].groupby(['DayOfWeek']).mean().plot()
plt.show()
```
